## 4. Data Cleaning & Preprocessing

```python
# Check missing values
df.isnull().sum()
```

### Observations

- The dataset contains missing values only in the **Review Rating** column.
- A total of **37 records** (less than 1% of the dataset) have missing review ratings.
- Since the missing proportion is very small, the values can be imputed rather than removing entire records.

### Handling Missing Values

The `Review Rating` column contains 37 missing values. Instead of replacing them with the overall median, the missing values will be imputed using the median review rating within each product category. This approach preserves category-specific rating patterns and reduces potential bias.

```python
# Fill missing review ratings using the median rating within each category
df["Review Rating"] = (
    df.groupby("Category")["Review Rating"]
      .transform(lambda x: x.fillna(x.median()))
)
```

```python
# Verify that all missing values have been handled
df.isnull().sum()
```

### Result

All missing values in the dataset have been successfully handled. The Review Rating column was imputed using the median rating within each product category, preserving category-level characteristics while maintaining all customer records.

## 5. Standardize Column Names

To improve code readability and maintain consistency across Python, SQL, and Power BI, all column names will be converted to lowercase and spaces will be replaced with underscores (snake_case naming convention).

```python
# Convert column names to lowercase and replace spaces with underscores
df.columns = (
    df.columns
      .str.lower()
      .str.replace(" ", "_")
)
```

```python
df.columns
```

```python
# Rename purchase_amount_(usd) to purchase_amount
df = df.rename(
    columns={
        "purchase_amount_(usd)": "purchase_amount"
    }
)
```

## 5. Feature Engineering

Create new variables that improve business analysis.

```python
# Create age groups

df["age_group"] = pd.cut(
    df["age"],
    bins=[17, 25, 35, 50, 100],
    labels=["18-25", "26-35", "36-50", "51+"]
)
```
